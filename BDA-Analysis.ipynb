**Session Analysis**

```python
import pandas as pd

# Convert 'TimeStamp' to datetime
data['TimeStamp'] = pd.to_datetime(data['TimeStamp'])

# Sort by IP (IpId) and timestamp
data = data.sort_values(by=['IpId', 'TimeStamp'])

# Calculate time difference between consecutive requests for each IP
data['TimeDiff'] = data.groupby('IpId')['TimeStamp'].diff()

# Define session threshold (2 mins minimum, 120 mins maximum)
session_min = pd.Timedelta(minutes=2)
session_max = pd.Timedelta(minutes=120)

# Start a new session if TimeDiff exceeds 30 minutes or is NaT
data['NewSession'] = (data['TimeDiff'] > pd.Timedelta(minutes=30)) | (data['TimeDiff'].isna())

# Assign a session ID
data['SessionId'] = data.groupby('IpId')['NewSession'].cumsum()

# Group sessions and filter by valid session duration
session_data = data.groupby(['IpId', 'SessionId']).agg(
    session_start=('TimeStamp', 'min'),
    session_end=('TimeStamp', 'max'),
    num_requests=('TimeStamp', 'count'),
    session_duration=('TimeDiff', lambda x: x.sum().total_seconds() / 60),  # in minutes
)

# Filter valid sessions
valid_sessions = session_data[
    (session_data['session_duration'] >= session_min.total_seconds() / 60) &
    (session_data['session_duration'] <= session_max.total_seconds() / 60)
].reset_index()

# Merge back with main data to analyze session details
session_details = data.merge(valid_sessions[['IpId', 'SessionId']], on=['IpId', 'SessionId'], how='inner')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

def analyze_session_patterns(session_details):
    """
    Analyze session patterns from the pre-processed session_details DataFrame
    """
    # Add time of day category
    session_details['TimeOfDay'] = pd.to_datetime(session_details['TimeStamp']).dt.hour.map(
        lambda x: 'Morning' if 5 <= x < 12
        else 'Afternoon' if 12 <= x < 17
        else 'Evening' if 17 <= x < 22
        else 'Night'
    )
    
    # 1. URI Type Analysis
    uri_analysis = session_details.groupby(['URI_Type']).agg({
        'SessionId': 'nunique',
        'IpId': 'count'
    }).rename(columns={
        'SessionId': 'Unique_Sessions',
        'IpId': 'Total_Visits'
    })
    uri_analysis['Avg_Visits_Per_Session'] = (
        uri_analysis['Total_Visits'] / uri_analysis['Unique_Sessions']
    ).round(2)
    
    # 2. Referrer Type Analysis
    referrer_analysis = session_details.groupby(['Referrer_Type']).agg({
        'SessionId': 'nunique',
        'IpId': 'count'
    }).rename(columns={
        'SessionId': 'Unique_Sessions',
        'IpId': 'Total_Visits'
    })
    referrer_analysis['Avg_Visits_Per_Session'] = (
        referrer_analysis['Total_Visits'] / referrer_analysis['Unique_Sessions']
    ).round(2)
    
    # 3. Country, Device, and Browser Analysis
    # Step 1: Aggregate data by country, device type, and browser
    analysis_columns = ['CountryCode', 'Device_Type', 'Browser']
    country_device_browser_analysis = session_details.groupby(analysis_columns).agg({
        'SessionId': 'nunique',
        'TimeDiff': lambda x: x.mean().total_seconds() / 60 if not x.empty else 0
    }).rename(columns={
        'SessionId': 'Number_of_Sessions',
        'TimeDiff': 'Avg_Time_Spent_Minutes'
    }).round(2)

    # Step 2: Calculate total sessions per country
    total_sessions_per_country = country_device_browser_analysis.groupby('CountryCode')['Number_of_Sessions'].sum()

    # Step 3: Identify the top 3 countries
    top_countries = total_sessions_per_country.nlargest(3).index

    # Step 4: Filter to include only top countries
    country_device_browser_analysis = country_device_browser_analysis.loc[top_countries]

    # 4. Session Journey Analysis
    session_journeys = session_details.groupby('SessionId').agg({
        'URI_Type': lambda x: list(x),
        'CountryCode': 'first',
        'Device_Type': 'first',
        'Browser': 'first',
        'TimeDiff': lambda x: x.sum().total_seconds() / 60
    }).rename(columns={'TimeDiff': 'Total_Session_Duration_Minutes'})
    
    return {
        'uri_analysis': uri_analysis,
        'referrer_analysis': referrer_analysis,
        'country_device_browser_analysis': country_device_browser_analysis,
        'session_journeys': session_journeys
    }

def create_visualizations(results):
    """
    Create visualizations for the analysis results
    """
    # Set style for better visualizations
    plt.style.use('seaborn-v0_8-whitegrid')
    
    # 1. URI Type Distribution
    plt.figure(figsize=(12, 6))
    results['uri_analysis']['Unique_Sessions'].plot(kind='bar', color='skyblue')
    plt.title('Number of Sessions by URI Type')
    plt.xlabel('URI Type')
    plt.ylabel('Number of Sessions')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # 2. Referrer Type Distribution
    plt.figure(figsize=(12, 6))
    results['referrer_analysis']['Unique_Sessions'].plot(kind='bar', color='coral')
    plt.title('Number of Sessions by Referrer Type')
    plt.xlabel('Referrer Type')
    plt.ylabel('Number of Sessions')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
    # 3. Country and Device Heatmap
    pivot_device_country = results['country_device_browser_analysis']['Avg_Time_Spent_Minutes'].unstack(['Device_Type'])
    plt.figure(figsize=(12, 8))
    sns.heatmap(pivot_device_country, annot=True, fmt='.1f', cmap='Blues', cbar=True)
    plt.title('Average Time Spent (Minutes) by Device Types and Countries')
    plt.tight_layout()
    plt.show()

    # 4. Country and Browser Heatmap
    pivot_browser_country = results['country_device_browser_analysis']['Avg_Time_Spent_Minutes'].unstack(['Browser'])
    plt.figure(figsize=(12, 8))
    sns.heatmap(pivot_browser_country, annot=True, fmt='.1f', cmap='Greens', cbar=True)
    plt.title('Average Time Spent (Minutes) by Browser Types and Countries')
    plt.tight_layout()
    plt.show()

def print_analysis_summary(results):
    """
    Print a summary of the analysis results
    """
    print("\n=== URI Type Analysis ===")
    print(results['uri_analysis'])
    
    print("\n=== Referrer Type Analysis ===")
    print(results['referrer_analysis'])
    
    print("\n=== Top Countries by Average Session Duration ===")
    country_summary = results['session_journeys'].groupby('CountryCode')[
        'Total_Session_Duration_Minutes'
    ].agg(['mean', 'count']).round(2)
    print(country_summary.sort_values('mean', ascending=False))
    
    print("\n=== Most Common Session Journeys ===")
    journey_patterns = results['session_journeys']['URI_Type'].apply(tuple).value_counts().head()
    print(journey_patterns)

def generate_full_analysis(session_details):
    """
    Generate and display complete analysis
    """
    results = analyze_session_patterns(session_details)
    create_visualizations(results)
    print_analysis_summary(results)
    return results
```
